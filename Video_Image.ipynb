#video to image Conversion(30 Frames Per Minutes)

```python
from google.colab import files
import cv2
import os

# Step 1: Upload video to Colab
uploaded = files.upload()  # Choose your video file when prompted
video_path = list(uploaded.keys())[0]  # Get the uploaded file's name

# Step 2: Set up output folder and frames per minute
output_folder = "extracted_frames"
frames_per_minute = 30  # Number of frames to extract per minute of video

# Step 3: Create output directory
os.makedirs(output_folder, exist_ok=True)

# Step 4: Extract frames
def video_to_images(video_path, output_folder, frames_per_minute):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)  # Frames per second
    if fps == 0:
        print("Error: Could not retrieve FPS from video.")
        return

    frame_interval = int(fps * 60 / frames_per_minute)  # Interval in frames
    frame_count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break  # Exit when the video ends

        # Save frame at specified intervals
        if frame_count % frame_interval == 0:
            filename = os.path.join(output_folder, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(filename, frame)
            saved_count += 1

        frame_count += 1

    cap.release()
    print(f"Extraction complete. {saved_count} frames saved to '{output_folder}'.")

# Call the function
video_to_images(video_path, output_folder, frames_per_minute)

# Step 5: Zip and download the extracted frames
!zip -r extracted_frames.zip extracted_frames
files.download("extracted_frames.zip")
```
